```python
import os
import json
import matplotlib.pyplot as plt

RESULTS_DIR = "../results"
```

```python
def load_metrics(label):
    path = os.path.join(RESULTS_DIR, f"{label}_metrics.json")
    with open(path, "r") as f:
        return json.load(f)

before = load_metrics("vm_before")
after = load_metrics("vm_after")
```

```python
cpu = [before["cpu_percent"], after["cpu_percent"]]
memory = [before["memory"]["percent"], after["memory"]["percent"]]
disk = [before["disk"]["percent"], after["disk"]["percent"]]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
labels = ["Antes", "Después"]

axs[0].bar(labels, cpu, color=["skyblue", "orange"])
axs[0].set_title("Uso de CPU (%)")
axs[0].set_ylim(0, 100)

axs[1].bar(labels, memory, color=["lightgreen", "red"])
axs[1].set_title("Uso de Memoria (%)")
axs[1].set_ylim(0, 100)

axs[2].bar(labels, disk, color=["purple", "gold"])
axs[2].set_title("Uso de Disco (%)")
axs[2].set_ylim(0, 100)

plt.suptitle("Comparación de Recursos - Antes y Después del Benchmark")
plt.tight_layout()
plt.show()
```

```python
def mostrar_txt(nombre):
    path = os.path.join(RESULTS_DIR, f"{nombre}.txt")
    with open(path, "r") as f:
        print(f.read())

mostrar_txt("cpu_test")
mostrar_txt("memory_test")
mostrar_txt("disk_test")
```
